```python
import sys
import pandas as pd
from unidecode import unidecode

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
url = 'https://www.bcb.gov.br/controleinflacao/historicotaxasjuros'
```

```python
s = Service('path/to/chrome-driver')
driver = webdriver.Chrome(service=s)
driver.get(url)
```

```python
nmClass = 'table-responsive'
elemento_tabela_selic = driver.find_element(By.CLASS_NAME, nmClass)
html_element_tab_titulos = elemento_tabela_selic.get_attribute('innerHTML')
```

```python
tables = pd.read_html(html_element_tab_titulos)
```

```python
df_selic = tables[0]
```

```python
df_selic.head()
```

```python
df_selic.columns = df_selic.columns.droplevel()
```

```python
df_selic.head()
```

```python
_columns = {
    'nº': 'reuniao_n'
    , 'data' : 'reuniao_data'
    , 'viés' : 'reuniao_vies'
    , 'Período de vigência' : 'periodo_vigencia'
    , 'Meta SELIC % a.a. (1)(6)' : 'meta_selic_aa'
    , 'TBAN % a.m. (2)(6)' : 'tban_am'
    , '% (3)' : 'taxa_selic_relativa'
    , '% a.a.(4)' : 'taxa_selic_aa'
}

df_selic = df_selic.rename(columns=_columns)
```

```python
df_selic.head()
```

```python
df_selic['reuniao_n'] = df_selic['reuniao_n'].apply(lambda x: str(x).replace('º', ''))
```

```python
df_selic.head()
```

```python
df_selic['periodo_vigencia_inicio'] = df_selic['periodo_vigencia'].apply(lambda x: str(x).split('-')[0].replace(' ', ''))
```

```python
df_selic['periodo_vigencia_fim'] = df_selic['periodo_vigencia'].apply(lambda x: str(x).split('-')[1].replace(' ', ''))
```

```python
df_selic.head()
```

```python
df_selic['meta_selic_aa'] = df_selic['meta_selic_aa'].apply(lambda x: float(x) / 100)
```

```python
df_selic['taxa_selic_relativa'] = df_selic['taxa_selic_relativa'].apply(lambda x: float(x) / 100)
```

```python
df_selic['taxa_selic_aa'] = df_selic['taxa_selic_aa'].apply(lambda x: float(x) / 100)
```

```python
df_selic.head()
```

```python
df_selic.to_csv('historico_taxa_selic.csv')
```
